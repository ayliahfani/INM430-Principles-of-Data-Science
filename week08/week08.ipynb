## Exercise 1: Lengthy Books
### find out which book as more words

```python
# Les Miserables
lesmis_url = 'https://www.gutenberg.org/files/1184/1184-0.txt'
lesmis = urlopen(lesmis_url).read().decode('utf-8')
lesmis_words = word_tokenize(lesmis)

# War and Peace
warandpeace_url = 'https://www.gutenberg.org/files/2600/2600-0.txt'
warandpeace = urlopen(warandpeace_url).read().decode('utf-8')
warandpeace_words = word_tokenize(warandpeace)

# Don Quijote
donquijote_url = 'http://www.gutenberg.org/cache/epub/2000/pg2000.txt'
donquijote = urlopen(donquijote_url).read().decode('utf-8')
donquijote_words = word_tokenize(donquijote)

print("les Miserable: " + str(len(lesmis_words)))
print("War and Peace: "+str(len(warandpeace_words)))
print("Don Quijote: "+str(len(donquijote_words)))
```

### Break the books into sentences

```python
lesmis_sentences = sent_tokenize(lesmis)
warandpeace_sentences = sent_tokenize(warandpeace)
donquijote_sentences = sent_tokenize(donquijote)
```

### Count how many sentences each book has

```python
print("les Miserables: "+str(len(lesmis_sentences)))
print("War and peace: "+str(len(warandpeace_sentences)))
print("Don Quijote: "+str(len(donquijote_sentences)))
```

### Find the average length of a sentence

```python
lm_sentence_length = 0
wp_sentence_length = 0
dq_sentence_length = 0
# Les Miserables
for i in range(len(lesmis_sentences)):
    sentence_lesmis = lesmis_sentences[i].split()
    lm_sentence_length = len(sentence_lesmis) + lm_sentence_length
avg_lesmis = lm_sentence_length/len(lesmis_sentences)
print("Les Miserable avg words per sentence: "+ str(int(avg_lesmis)))

# War and Peace
for i in range(len(warandpeace_sentences)):
    sentence_wp = warandpeace_sentences[i].split()
    wp_sentence_length = len(sentence_wp) + wp_sentence_length
avg_warandpeace = wp_sentence_length/len(warandpeace_sentences)
print("War and Peace avg words per sentence: "+ str(int(avg_warandpeace)))

# Don Quijote
for i in range(len(donquijote_sentences)):
    sentence_dq = donquijote_sentences[i].split()
    dq_sentence_length = len(sentence_dq) + dq_sentence_length
avg_donquijote = dq_sentence_length/len(donquijote_sentences)
print("Don Quijote avg words per sentence: "+ str(int(avg_donquijote)))
```
